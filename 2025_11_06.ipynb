```python
from goose3 import Goose
article_url='https://www.iuj.ac.jp/about-2-2/principles/'
article=Goose().extract(article_url)
print(article.cleaned_text)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
sentence_tokens = sent_tokenize(article.cleaned_text)
print(len(sentence_tokens))
```

```python
result=[]
for s in sentence_tokens:
  result.append(sent(s).get('compound'))
print(result)
```

```python
import pandas as pd
output=pd.DataFrame(result, columns=['compound'])
if output.compound.mean()>0:
  print("The article is positive")
else:
  print("The article is negative")
```

```python
import pandas as pd
output=pd.DataFrame(result, columns=['compound'])
print(output.mean()[0])
if output.mean()[0]>0:
  print("The article is positive")
else:
  print("The article is negative")
```
