```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
import io

# 1. Upload the file and load it into a Pandas DataFrame
# You will be prompted to select and upload 'diamonds.csv'
print("Please upload your 'diamonds.csv' file when prompted:")
uploaded = files.upload()

# Read the uploaded CSV file
df = pd.read_csv(io.StringIO(uploaded['diamonds.csv'].decode('utf-8')))

# Display the first few rows and column information
print("\nFirst 5 rows of the dataset:")
print(df.head())
print("\nData information:")
df.info()
```

```python
# a. Get unique values and their counts in each column.
print("\n--- Question (a): Unique Values and Counts ---")
for col in df.columns:
    print(f"\nColumn: {col}")
    # Display unique values
    print("Unique Values:", df[col].unique())

    # Display value counts (for categorical columns, first 10 for continuous)
    if df[col].dtype == 'object' or len(df[col].unique()) < 20:
        print("Value Counts:\n", df[col].value_counts())
    else:
        print(f"Top 10 Value Counts (out of {df[col].nunique()} unique values):\n", df[col].value_counts().head(10))
```

```python
# b. What are the highest/lowest price of diamonds?
print("\n--- Question (b): Highest/Lowest Price ---")
highest_price = df['price'].max()
lowest_price = df['price'].min()
print(f"Highest Price: ${highest_price}")
print(f"Lowest Price: ${lowest_price}")
```

```python
# c. Create a variable called size = x*y*z.
df['size'] = df['x'] * df['y'] * df['z']
print("\n--- Question (c): 'size' variable created (volume in mm³) ---")
print("First 5 'size' values:\n", df['size'].head())

# d. Calculate the correlation of size and carat.
correlation_size_carat = df['size'].corr(df['carat'])
print("\n--- Question (d): Correlation of 'size' and 'carat' ---")
print(f"Correlation between size (x*y*z) and carat: {correlation_size_carat:.4f}")
```

```python
# e. Calculate the average prices in different clarities, cut, and color.
print("\n--- Question (e): Average Prices by Category ---")

avg_price_clarity = df.groupby('clarity')['price'].mean().sort_values(ascending=False)
print("\nAverage Price by Clarity (Highest to Lowest):\n", avg_price_clarity)

avg_price_cut = df.groupby('cut')['price'].mean().sort_values(ascending=False)
print("\nAverage Price by Cut (Highest to Lowest):\n", avg_price_cut)

avg_price_color = df.groupby('color')['price'].mean().sort_values(ascending=False)
print("\nAverage Price by Color (Highest to Lowest):\n", avg_price_color)
```

```python
# f. Plot a scatter plot for the x-axis=carat and y-axis=price.
print("\n--- Question (f): Scatter Plot (carat vs price) ---")

# Set the plot style
sns.set_style("whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df, alpha=0.5, hue='cut', palette='viridis')
plt.title('Diamond Price vs. Carat Weight, colored by Cut')
plt.xlabel('Carat')
plt.ylabel('Price (US$)')
plt.show()
```

```python
# g. 1) Convert the categorical variable 'color' as {'E':1, 'I':2, 'J':3, 'H':4, 'F':5, 'G':6, 'D':7}
color_mapping = {'E': 1, 'I': 2, 'J': 3, 'H': 4, 'F': 5, 'G': 6, 'D': 7}
df['color_encoded'] = df['color'].map(color_mapping)

print("\n--- Question (g.1): Categorical Encoding for 'color' ---")
print("Color Mapping Used:", color_mapping)
print("\nFirst 5 'color' and 'color_encoded' values:")
print(df[['color', 'color_encoded']].head())
```

```python
# Step 12 (h): OLS regression with ln(price)
import statsmodels.api as sm

# Prepare data
df = df.copy()
df['ln_price'] = np.log(df['price'])

# Encode 'cut' and 'clarity' as numerical variables
cut_mapping = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
clarity_mapping = {'I1': 1, 'SI2': 2, 'SI1': 3, 'VS2': 4, 'VS1': 5, 'VVS2': 6, 'VVS1': 7, 'IF': 8}

df['cut_num'] = df['cut'].map(cut_mapping)
df['clarity_num'] = df['clarity'].map(clarity_mapping)

# 'color_encoded' is already created in a previous step
# Rename 'color_encoded' to 'color_num' for consistency with the error message
df = df.rename(columns={'color_encoded': 'color_num'})


predictors = ['carat','cut_num','color_num','clarity_num','depth','table','size']
X = df[predictors].astype(float)
X = sm.add_constant(X)  # add intercept
y = df['ln_price'].astype(float)

# Fit OLS
model = sm.OLS(y, X, missing='drop').fit()
print(model.summary())

# Save regression table (coefficients, p-values) to a DataFrame
reg_table = model.summary2().tables[1].reset_index().rename(columns={'index':'variable'})
display(reg_table)
```

```python
# Step 13 (i): Formal interpretation guidance and automatic extraction of top effects
# Show key fit statistics:
r2 = model.rsquared
adj_r2 = model.rsquared_adj
n_obs = int(model.nobs)
print(f"Model R-squared = {r2:.4f}, Adjusted R-squared = {adj_r2:.4f}, Observations = {n_obs}\n")

# Extract coefficients and significance
coef_df = reg_table.copy()
coef_df['abs_t'] = coef_df['t'].abs()
# Sort by absolute t-value (statistical importance) and by absolute coefficient (economic magnitude)
by_t = coef_df.sort_values('abs_t', ascending=False)
by_coef = coef_df.sort_values(by='Coef.', key=lambda s: s.abs(), ascending=False)

print("Top predictors by statistical importance (abs t):")
display(by_t.head(6))

print("Top predictors by coefficient magnitude (abs coef):")
display(by_coef.head(6))

# Formal interpretation template (fill-in)
print("\nFormal interpretation template (academic):")
print("""
Dependent variable: ln(price). Coefficients approximate semi-elasticities:
- For a one-unit increase in predictor X_j (holding other variables constant), ln(price) changes by beta_j units.
  Approximately, the price changes by (exp(beta_j)-1)*100 % for larger betas, or ≈100 * beta_j percent for small beta_j.

Key points to check:
1. Coefficient sign and economic meaning (positive/negative).
2. Statistical significance (p-value < 0.05 indicates significance at 5% level).
3. Magnitude: compare coefficients scaled to meaningful units (e.g., 1 carat change).
4. Multicollinearity: check VIFs (see previous step). If VIF > 10 consider dropping or transforming variables.

Example sentence (fill with actual numbers from table):
'Controlling for other variables, a one-carat increase is associated with approximately β_carat (p = p_carat) change in ln(price),
which implies about (exp(β_carat)-1)*100 % change in price. The coefficient is (statistically significant / not statistically significant) at the 5% level.'
""")
```
